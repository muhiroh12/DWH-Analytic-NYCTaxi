```python
import pandas as pd
from skimpy import skim, generate_test_data
#from pandas_profiling import ProfileReport
```

```python
year='2023'
year_nt = int(year)
month='01'
month_nt = int(month)
path_raw=f"D:/zoomcamp/green_taxi/include/dataset/raw/nyc_taxi"
filename=f"green_tripdata_{year}_{month}.parquet"
path_transform = f"D:/zoomcamp/green_taxi/include/dataset/transform"
filename_tf = f"green_tripdata_{year}_{month}.csv"
```

```python
df = pd.read_parquet(path_raw + '/' + filename)
```

```python
skim(df)
```

```python
df = df.dropna(subset=['passenger_count', 'trip_distance', 'fare_amount'])
```

```python
df = df[df['trip_distance'] > 0]
```

```python
df = df[df['passenger_count'] > 0]
```

```python
df = df[df['fare_amount'] > 0]
df = df[df['total_amount'] > 0]
```

```python
df.head()
```

```python
df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime'], errors='coerce')
df['lpep_dropoff_datetime'] = pd.to_datetime(df['lpep_dropoff_datetime'], errors='coerce')

df = df[df['lpep_pickup_datetime'] < df['lpep_dropoff_datetime']]
```

```python
# Filter data untuk hanya mengambil tahun 2023
df_filtered = df[df['lpep_pickup_datetime'].dt.year == year_nt]
```

```python
df_filtered = df_filtered[df_filtered['lpep_pickup_datetime'].dt.month == month_nt]
```

```python
df_filtered.to_csv(path_transform + '/' + filename_tf, index=False)
```
